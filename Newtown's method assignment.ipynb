```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
y1=np.exp(-x/5)
y2=np.sin(x)
plt.plot(x,y1,'r',x,y2,'g')
plt.legend(['exp(x)','sin(x)'])
```

```python
def f(x):
    return np.exp(-x/5)-np.sin(x)
print("Enter the values for a and b")
for i in range(4):
    a = input()
    a = int(a)
    b = input()
    b = int(b)
    c = (a+b)/2
    tol = 10**(-7)
    i = 0
    while(abs(f(c))>=tol):
        if f(a)*f(c)<0:
            b = c
        else:
            a = c
        c = (a+b)/2
        i = i+1
    print('Approximate value =',c,' No of iteration =',i)
```

```python
def f(x):
    return np.exp(-x/5)-np.sin(x)
print("Enter values for a and b")
for i in range(4):
    a = input()
    a = int(a)
    b = input()
    b = int(b)
    c = (a*f(b)-b*f(a))/(f(b)-f(a))
    tol = 10**(-7)
    i = 0
    while(abs(f(c))>=tol):
        if f(a)*f(c)<0:
            b = c
        else:
            a = c
        c = (a*f(b)-b*f(a))/(f(b)-f(a))
        i = i+1
    print('Approximate value =',c,' No of iteration =',i)
```

```python
def f(x):
    return np.exp(-x/5)-np.sin(x)
def df(x):
    return (-1/5)*np.exp(-x/5)-np.cos(x)
print("Enter value for x0")
for i in range(4):
    x0 = input()
    x0 = int(x0)
    tol = 10**(-7)
    i = 0
    while abs(f(x0))>=tol:
        x1 = x0-f(x0)/df(x0)
        x0 = x1
        i = i+1
    print('Approximate value =',x1,' No of iteration =',i)
```

```python
def f(x):
    return 2**(-x)-np.cos(x)
print("Enter values for a and b")
for i in range(3):
    a = input()
    a = int(a)
    b = input()
    b = int(b)
    c = (a*f(b)-b*f(a))/(f(b)-f(a))
    tol = 10**(-7)
    i = 0
    while(abs(f(c))>=tol):
        if f(a)*f(c)<0:
            b = c
        else:
            a = c 
        c = (a*f(b)-b*f(a))/(f(b)-f(a))
        i = i+1
    print('Approximate value =',c,' No of iteration =',i)
```

Speed of convergence of Regula-Falsi method is greater than that of bisection method

```python
def f(x):
    return 2**(-x)-np.cos(x)
def df(x):
    return -2**(-x)*np.log(2) + np.sin(x)
print("Enter values for x0")
for i in range(3):
    x0=input()
    x0=int(x0)
    tol=10**(-7)
    i=0
    while abs(f(x0))>=tol:
        x1=x0-f(x0)/df(x0)
        x0=x1
        i=i+1
    print('Approximate value=',x1,' No of iteration=',i)
```

Newton's method > Regula-Falsi method > Bisection Method
